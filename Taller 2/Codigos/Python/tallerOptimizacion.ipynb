Punto 1 - Mezcla de fertilizantes




```python
import numpy as np
from scipy.optimize import minimize

"""Primero vamos a plantear los recursos que tenemos"""
A_disponible = 120
B_disponible = 80
C_disponible = 150
"""Ahora vamos a plantear las ganancias que deja cada fertilizante"""
ganancia_F1 = 180
ganancia_F2 = 220

"""ahora planteamos la funcion, pero como scypy minimiza la planteamos de manera negativa"""
def funcion_objetivo(x):
    x1, x2 = x[0], x[1]
    return -(ganancia_F1 * x1 + ganancia_F2 * x2)

"""Planteamos las retricciones despejando el maxinmo de materia dejando toda la desigualdad en una ecuacion"""

def restriccion_A(x):
    x1, x2 = x[0], x[1]
    return A_disponible - (3*x1 + 4*x2)

def restriccion_B(x):
    x1, x2 = x[0], x[1]
    return B_disponible - (2*x1 + 1*x2)

def restriccion_C(x):
    x1, x2 = x[0], x[1]
    return C_disponible - (5*x1 + 3*x2)

"""Danmos las caracteristicas de las restricciones"""
x0 = [10, 10]

restricciones = [
    {'type': 'ineq', 'fun': restriccion_A},
    {'type': 'ineq', 'fun': restriccion_B},
    {'type': 'ineq', 'fun': restriccion_C}
]

"""No negatividad"""
limites = [(0, None), (0, None)]

resultado = minimize(
    funcion_objetivo,
    x0,
    method='SLSQP',
    bounds=limites,
    constraints=restricciones
)

"""solucion"""
x1_opt, x2_opt = resultado.x
ganancia_max = -resultado.fun

"""Uso de recursos"""
uso_A = 3*x1_opt + 4*x2_opt
uso_B = 2*x1_opt + 1*x2_opt
uso_C = 5*x1_opt + 3*x2_opt

print(" Mezcla de fertilizantes")
print()
print(f"Fertilizante F1 (x1): {x1_opt:.4f} ton/semana")
print(f"Fertilizante F2 (x2): {x2_opt:.4f} ton/semana")
print()
print(f"Ganancia máxima: ${ganancia_max:.2f}")
print()
print("Uso de compuestos A, B, C")
print(f"Compuesto A usado: {uso_A:.4f} kg, Disponible: {A_disponible} kg")
print(f"Compuesto B usado: {uso_B:.4f} kg, Disponible: {B_disponible} kg")
print(f"Compuesto C usado: {uso_C:.4f} kg, DIsponible: {C_disponible} kg")
```
